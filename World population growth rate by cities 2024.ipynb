```python
#This dataset was downloaded from Kaggle 
#About this dataset : The below dataset provides the growth rate of the population of the top 800 populated cities around the world for 2023 and 2024. 
#This dataset provides key information, such as the name of the city, the country in which that city is located in, the continent it belongs to, and the populations for the years 2023 and 2024. 
#Additionally, and the last column is about the annual growth rate for each city.
```

```python
#Check data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


file_path = '/Users/90001202/Desktop/Wprld population growth rate by cities 2024.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
#Average Growth Rate by continent

C_avr=df.groupby('Continent').agg({'Growth Rate':'mean'}).sort_values(by='Growth Rate',ascending=False).reset_index()
print(C_avr)

sns.barplot(x='Continent',y='Growth Rate', data=C_avr,palette='Blues_d')
plt.xticks(rotation=45)
plt.show() 

import warnings
warnings.filterwarnings('ignore')

##We can see that Afirca has the highest growth rate while Europe has the lowest
```

```python
#Analyze distribution of Growth Rate

plt.figure(figsize=(10,6))
sns.displot(df['Growth Rate'],kde=True)
plt.title('Distribution of Growth Rate')
plt.xlabel('Growth Rate')
plt.show

#Analyze distribution of Growth rate by continent
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Continent', y='Growth Rate', palette='Set1')
plt.show
```

```python
# Top 10 Countries of Growth Rate
country_top_10=df.groupby('Country')['Growth Rate'].mean().sort_values(ascending=False).head(10).reset_index()
print(country_top_10)

sns.barplot(data=country_top_10, x='Growth Rate', y='Country',  palette='Set1')
plt.title('Top 10 contries by Growth Rate')
plt.xlabel('Growth Rate')
plt.ylabel('Country')
plt.show
```

```python
# Top 10 city by Growth Rate
city_top_10=df.groupby('City')['Growth Rate'].mean().sort_values(ascending=False).head(10).reset_index()
print(city_top_10)

sns.barplot(data=city_top_10, x='Growth Rate', y='City',  palette='Set1')
plt.title('Top 10 Cities by Growth Rate')
plt.xlabel('Growth Rate')
plt.ylabel('City')
plt.show
```

```python
#Bottom 10 country by Growth Rate

country_b_10=df.groupby('Country')['Growth Rate'].mean().sort_values(ascending=False).tail(10).reset_index()
print(country_b_10)

sns.barplot(data=country_b_10, x='Growth Rate', y='Country',  palette='Set2')
plt.title('Bottom 10 contries by Growth Rate')
plt.xlabel('Growth Rate')
plt.ylabel('Country')
plt.show
```

```python
#Bottom 10 city by Growth Rate

city_b_10=df.groupby('City')['Growth Rate'].mean().sort_values(ascending=False).tail(10).reset_index()
print(city_b_10)

sns.barplot(data=city_b_10, x='Growth Rate', y='City',  palette='Set2')
plt.title('Bottom 10 cities by Growth Rate')
plt.xlabel('Growth Rate')
plt.ylabel('City')
plt.show
```

```python
#2024 Top 10 Country by population
country_top_10_po=df.groupby('Country')['Population (2024)'].sum().sort_values(ascending=False).head(10).reset_index()
print(country_top_10_po)

sns.barplot(data=country_top_10_po, x='Population (2024)', y='Country',  palette='Set3')
plt.title('2024 Top 10 contries by population')
plt.xlabel('Population')
plt.ylabel('Country')
plt.show
```

```python
#2024 Top 10 City by population
city_top_10_po=df.groupby('City')['Population (2024)'].sum().sort_values(ascending=False).head(10).reset_index()
print(country_top_10_po)

sns.barplot(data=city_top_10_po, x='Population (2024)', y='City',  palette='Set3')
plt.title('2024 Top 10 city by population')
plt.xlabel('Population')
plt.ylabel('City')
plt.show
```
